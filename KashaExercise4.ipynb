```python
#  Imports and load data
%pip install openpyxl       # needed for pandas to read .xlsx files
import pandas as pd
import matplotlib.pyplot as plt

def display_dataframe(name, df):
    print(f"\n=== {name} ===")
    display(df.head(10))


xlsx_path = "kasha_data/exercise_4_agile.xlsx"
df_prod = pd.read_excel(xlsx_path, sheet_name='Product Backlog')
df = pd.read_excel(xlsx_path, sheet_name='Sprint Backlog')   # work from sprint sheet for sprint-level metrics

# normalize column names (trim)
df.columns = [c.strip() for c in df.columns]
df_prod.columns = [c.strip() for c in df_prod.columns]

print("Sprint Backlog columns:", df.columns.tolist())
print("Product Backlog columns:", df_prod.columns.tolist()) 

# Quick head preview 
display_dataframe("Sprint Backlog sample", df.head(10))
display_dataframe("Product Backlog sample", df_prod.head(10))
```

```python
# helper function to get logically sorted sprints and choose last 8
def sprint_sort_key(s):
    # expects format like "Sprint YYYY-MM-N" -> returns (year, month, seq)
    parts = str(s).replace('Sprint ','').split('-')
    try:
        return (int(parts[0]), int(parts[1]), int(parts[2]))
    except:
        return (0,0,0)

sprints = sorted(df['Sprint'].dropna().unique(), key=sprint_sort_key)
last8 = sprints[-8:]
print("All sprints count:", len(sprints))
print("Last 8 sprints in order:", last8)
```

### **Velocity & Commitment metrics (planned vs completed)**

```python

# Planned: number of stories assigned to the sprint
# Completed: stories with Demo Status == 'Accepted'

planned_series = df[df['Sprint'].isin(last8)].groupby('Sprint').size()
completed_series = df[(df['Sprint'].isin(last8)) & (df['Demo Status']=='Accepted')].groupby('Sprint').size()

# Reindex to ensure last8 ordering, fill missing with 0
planned = planned_series.reindex(last8).fillna(0).astype(int)
completed = completed_series.reindex(last8).fillna(0).astype(int)

metrics = pd.DataFrame({
    'sprint': last8,
    'planned': planned.values,
    'completed': completed.values
})
metrics['commitment_ratio'] = (metrics['completed'] / metrics['planned']).round(3)
metrics['commitment_ratio'] = metrics['commitment_ratio'].fillna(0)  # handle planned=0

display_dataframe("Velocity & Commitment (last 8 sprints)", metrics)
```

**Plot planned vs completed and commitment ratio**

```python

# Ensure matplotlib uses the notebook backend
%matplotlib inline

fig, ax = plt.subplots(figsize=(11,5))
x = range(len(metrics))
ax.bar(x, metrics['planned'], label='Planned', color="#E5007D")
ax.bar(x, metrics['completed'], label='Completed', color="#333333")
ax.set_xticks(x); ax.set_xticklabels(metrics['sprint'], rotation=45, ha='right')
ax.set_ylabel('Number of Stories')
ax.set_title('Planned vs Completed (Last 8 Sprints)')
ax.legend()

ax2 = ax.twinx()
ax2.plot(x, metrics['commitment_ratio'], marker='o', linestyle='-', label='Commitment Ratio')
ax2.set_ylabel('Commitment Ratio (Completed / Planned)')
ax2.set_ylim(0,1.05)
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

### **Story Type Analysis (only Accepted stories)**

```python

accepted = df[df['Demo Status']=='Accepted'].copy()
story_type_counts = accepted['User Story Type'].value_counts()
story_type_pct  = accepted['User Story Type'].value_counts(normalize=True).mul(100).round(1)

story_type_summary = pd.DataFrame({
    'count': story_type_counts,
    'percent': story_type_pct
}).reset_index().rename(columns={'index':'User Story Type'})

display_dataframe("Story Type Composition (Accepted stories)", story_type_summary)
```

```python

plt.figure(figsize=(6,6))
plt.pie(
    story_type_summary['count'],
    labels=story_type_summary['User Story Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Set2.colors
)
plt.title("User Story Type Distribution (Accepted Stories)")
plt.show()
```

**Key Observations**

    Balanced Delivery Mix:

-Out of all accepted stories, 41.5% are Features, 39.5% are Support tasks, 15.2% are Defects, and 3.8% are Enablers.

-This indicates that the team is not only delivering new business functionality (Features) but also dedicating a significant amount of effort to support and maintenance work.

    High Share of Support Work:

-Nearly two-fifths of accepted stories are Support-related.While this helps ensure system stability and user satisfaction, a consistently high proportion of support tasks may reduce the teams capacity to deliver new value.

-It may suggest reactive work or operational issues that need attention, such as better automation, documentation, or issue prevention.

    Moderate Defect Ratio (15.2%):

-Around one in six accepted stories are Defects.

-This is within a reasonable range but should be monitored  if this trend grows, it could signal quality control or testing gaps earlier in the development cycle.

    Low Enabler Work (3.8%):

-A small percentage of Enabler stories suggests limited focus on technical debt reduction, performance improvements, or infrastructure readiness.

-Over time, underinvestment in enablers can slow down delivery speed or lead to scalability and maintainability issues.

**Effort comparison: avg Mandays (or Estimates) by User Story Type - accepted only**

```python

# pick effort column available
effort_col = None
for col in ['Mandays','Estimates','Est.','Estimate','Story Points']:
    if col in df.columns:
        effort_col = col
        break

if effort_col:
    effort_by_type = accepted.groupby('User Story Type')[effort_col].mean().round(2).reset_index().rename(columns={effort_col:'avg_effort'})
    display_dataframe(f"Average Effort ({effort_col}) by User Story Type (Accepted)", effort_by_type)
    # Summary specifically for Feature vs Defect vs Enabler
    types_of_interest = ['Feature','Defect','Enabler']
    filtered = accepted[accepted['User Story Type'].isin(types_of_interest)]
    summary = filtered.groupby('User Story Type').agg(count=('User Story Type','size'), avg_effort=(effort_col,'mean')).round(2).reset_index()
    display_dataframe("Feature vs Defect vs Enabler - Accepted stories", summary)
else:
    print("No effort column found (Mandays/Estimates). If you have an effort column, rename it to 'Mandays' or 'Estimates'.")
```

    Solution: Use backlog prioritization sessions with business stakeholders to balance operational stability (control/security) with growth outcomes (revenue/cost impact).

1. #### ***Flow Efficiency***

Definition:
Measures the ratio of active work time (time spent actually working on a story) to the total cycle time (including waiting or blocked time).

$$
\text{Flow Efficiency} = \dfrac{\text{Active Time}}{\text{Total Cycle Time}} \times 100
$$


Relevance?:

It highlights inefficiencies in the development workflow  e.g., long waiting times for reviews, dependencies, or approvals.

A low flow efficiency (<40%) indicates process bottlenecks or coordination issues, helping teams focus on reducing handoff delays and increasing throughput.

Clinically relevant: for a healthcare tech team, improving flow efficiency means faster turnaround for patient-impacting features and reduced time-to-value.


2. #### ***Escaped Defect Rate***

Definition:
The percentage of defects found after release compared to total defects discovered (pre- and post-release).

$$
\text{Escaped Defect Rate} = \dfrac{\text{Post-Release Defects}}{\text{Total Defects}} \times 100
$$


Relevance?:

Directly measures product quality and the effectiveness of testing and QA practices.

A high escaped defect rate indicates issues are being found by end users rather than internal testing, signaling a need for stronger test automation or earlier validation.

For clinical or healthcare platforms, reducing escaped defects is critical  it ensures system reliability, patient safety, and data accuracy.
